```python
def rolling_window(a, window):
    '''
    Take in an array and return array of rolling windows of specified length
    
    Parameters:
    - a: numpy array that will be windowed
    - window: integer that will be the length of the window
    
    Returns:
    - a_windowed: array where each entry is an array of length window
    '''
    shape = a.shape[:-1] + (a.shape[-1] - window + 1, window)
    strides = a.strides + (a.strides[-1],)
    a_windowed = np.lib.stride_tricks.as_strided(a, shape=shape, strides=strides)
    return a_windowed
```

```python
def time_series_to_point_cloud(time_series, dimension_embed=2):
    '''
    Convert a time series into a point cloud in the dimension specified by dimension_embed
    
    Parameters:
    - time_series: numpy array of time series values
    - dimension_embed: dimension of Euclidean space in which to embed the time series into by taking windows of dimension_embed length,
        e.g. if the time series is [t_1,...,t_n] and dimension_embed is 2, then the point cloud would be [(t_0, t_1), (t_1, t_2),...,(t_(n-1), t_n)]
        
    Returns:
    - point_cloud_normalized: point cloud normalized to take values between [0, 1]
    '''

    assert len(time_series) >= dimension_embed, 'dimension_embed larger than length of time_series'
    
    # time series to float type
    time_series = time_series.astype(float)

    # compute point cloud
    point_cloud = rolling_window(a=time_series, window=dimension_embed)
    
    return np.array(point_cloud)
```

```python
def point_cloud_to_persistent_homology(point_cloud, epsilon_range, max_simplex_dim):
    # take unique points in the point cloud only
    point_cloud = np.unique(point_cloud, axis=0)

    # build filtration with fill_rips
    filtration = d.fill_rips(point_cloud, max_simplex_dim, 1.0)
    
    # calculate persistent homology
    persistence = d.homology_persistence(filtration)
    
    # initialize persistence diagrams
    diagrams = d.init_diagrams(persistence, filtration)
    
    ep_ran_len = len(epsilon_range)
    
    homology = {dimension: np.zeros(ep_ran_len) for dimension in range(max_simplex_dim)}

    for dimension, diagram in enumerate(diagrams):
        if dimension < max_simplex_dim:
            homology[dimension] = np.array([np.array(((epsilon_range >= point.birth) & (epsilon_range <= point.death)).astype(int)) 
                                        for point in diagram
                                       ]).sum(axis=0)
        
    return homology
```

```python
def time_series_to_persistent_homology(time_series, epsilon_range, max_simplex_dim):
    '''
    Wrapper function that takes in a time series and outputs
    the persistent homology object, along with other
    auxiliary objects.
    
    Parameters: 
    - time_series: Numpy array of time series values
    - num_epsilon_steps: Either a positive integer or a list of epsilon values. If an integer, it is the number of segments the inverval [0, 1] 
    will be divided into. The method parameter will determine how the division will be set.
    - max_simplex_dim: Integer denoting the maximum dimension of simplexes to create in filtration
    
    Returns:
    '''
    
    # create point cloud from time series
    point_cloud = time_series_to_point_cloud(time_series, dimension_embed=max_simplex_dim)
    
#     # take unique points in the point cloud only
#     point_cloud = np.unique(point_cloud, axis=0)

#     # build filtration with fill_rips
#     filtration = d.fill_rips(point_cloud, max_simplex_dim, 1.0)
    
#     # calculate persistent homology
#     persistence = d.homology_persistence(filtration)
    
#     # initialize persistence diagrams
#     diagrams = d.init_diagrams(persistence, filtration)
    
#     homology = {}

#     for dimension in range(max_simplex_dim):
#         homology[dimension] = np.array([np.array(((epsilon_range >= point.birth) & (epsilon_range <= point.death)).astype(int)) 
#                                         for point in diagrams[dimension]
#                                        ]).sum(axis=0)
        
#     return homology

    homology = point_cloud_to_persistent_homology(point_cloud=point_cloud, epsilon_range=epsilon_range, max_simplex_dim=max_simplex_dim)
    return homology
```

```python
def betti(time_series, epsilon_range, max_simplex_dim, agg=None):
    homology = time_series_to_persistent_homology(time_series=time_series,
                                                  epsilon_range=epsilon_range,
                                                  max_simplex_dim=max_simplex_dim)
    
    if agg == 'sum':
        return np.array(homology.values()).sum().sum()
    elif agg == 'mean':
        return np.array(homology.values()).mean()
    elif agg == 'median':
        return np.median(np.array(homology.values()))
    else:
        return np.array(homology.values())
```

```python
def time_series_rolling_betti(time_series, epsilon_range, max_simplex_dim, window, agg=None):
    betti_results = [betti(time_series=wdw,
                       epsilon_range=epsilon_range,
                       max_simplex_dim=max_simplex_dim,
                       agg=agg) for wdw in rolling_window(time_series, window)]
    
    return pd.concat([time_series, pd.Series(data=betti_results, index=time_series.index[window-1:], name='betti')], axis=1)
```

```python
def simplex_conditions(row, columns, n):
    '''
    Helper function to apply conditions that single 
    out a simplex. To be used applied to a pandas DataFrame.
    The columns are assumed to be {1, 2,..., n, (1, 2), (1, 3), ...} where 
    n is the dimension of simplex being checked for. The columns 1, ..., n
    are assumed to be integer type and the columns of the form (i, j) are 
    assumed to be tuple type, and to have i < j.
    
    Parameters:
    - row: row of the DataFrame
    - columns: list of the columns of the DataFrame
    - n: dimnesion of the simplex being checked for
    
    Returns:
    - Boolean based on whether the row in question satisfies simplex conditions
    '''
    
    for col in columns:
        if type(col) == int and col < n:
            if row[col] == row[col+1]:
                return False
            
        elif type(col) == tuple:
            if row[col[1]] != row[col]:
                return False
            
    else:
        return True
```

```python
def create_n_simplexes(one_spxs, n):
    '''
    Helper function that creates an array of the 
    n simplexes from the data frame of the 
    one simplexes
    
    Parameters:
    - one_spxs: DataFrame of one simplexes 
    - n: dimension of simplexes being created
    
    Returns:
    - list of n+1-tuples that represent n simplexes
    '''
    
    # create copy of one simplexes
    df = one_spxs
    
    # iteratively merge on one simplexes
    for combo in list(itertools.combinations(range(n+1), 2))[1:]:
        if combo[0] == 0:
            df = pd.merge(
                left=df,
                right=one_spxs.rename(columns={1: combo[1]}),
                on=0,
                how='left'
            ).dropna().drop_duplicates().astype(int)
        else:
            df = pd.merge(
                left=df,
                right=one_spxs.rename(columns={0: combo[0], 1: combo}),
                on=combo[0],
                how='left'
            ).dropna().drop_duplicates().astype(int)
    
    # extract vertexes that make up the n simplexes
    cols = df.columns

    n_simplexes = []
    if len(df) > 0:
        n_simplexes = [tuple(simplex) 
                       for simplex in df[df.apply(lambda x: simplex_conditions(x, cols, n), axis=1)][[col 
                                                                                                      for col in df.columns if type(col) == int]].values]
    
    return n_simplexes
```

```python
def plot_two_simplexes(point_cloud, distance_matrix, epsilon, one_simplexes_color='#6DA5BA', two_simplexes_color='#E0513B'):
    one_simplicies = [(x,y) for (x,y) in zip(*np.where(distance_matrix <= epsilon)) if x < y]
    two_simplicies = create_n_simplexes(one_spxs=pd.DataFrame(one_simplicies), n=2)

    lines = [(point_cloud[x], point_cloud[y]) for (x, y) in one_simplicies]

    fig, ax = plt.subplots(figsize=(7, 7))

    lc2 = mc.LineCollection(
        lines,
        colors=one_simplexes_color,
        linewidths=0.5
    );

    pd.DataFrame(point_cloud).rename(columns={0: 'x', 1: 'y'}).plot(
        kind='scatter',
        x='x',
        y='y',
        ax=ax,
        s=7,
        c=one_simplexes_color
    );

    ax.add_collection(lc2);
    ax.autoscale();

    for two_simplex in two_simplicies:
        Path = mpath.Path
        path_data = [
                (Path.MOVETO, point_cloud[two_simplex[0]]),
                (Path.LINETO, point_cloud[two_simplex[1]]),
                (Path.LINETO, point_cloud[two_simplex[2]]),
                (Path.CLOSEPOLY, point_cloud[two_simplex[0]]),
                ]
        codes, verts = zip(*path_data)
        path = mpath.Path(verts, codes)
        patch = mpatches.PathPatch(path, facecolor=two_simplexes_color, alpha=0.2);
        ax.add_patch(patch);

    plt.show();
```

```python
def plot_two_simplexes_multi(point_cloud, distance_matrix, epsilon_list, one_simplexes_color='#6DA5BA', two_simplexes_color='#E0513B'):
    lgth = len(epsilon_list)
    
    fig, axs = plt.subplots(1,lgth, figsize=(7*lgth, 7))
    
    for idx, epsilon in enumerate(epsilon_list):
        one_simplicies = [(x,y) for (x,y) in zip(*np.where(distance_matrix <= epsilon)) if x < y]
        two_simplicies = create_n_simplexes(one_spxs=pd.DataFrame(one_simplicies), n=2)

        lines = [(point_cloud[x], point_cloud[y]) for (x, y) in one_simplicies]

    #     fig, ax = plt.subplots(figsize=(7, 7))

        lc2 = mc.LineCollection(
            lines,
            colors=one_simplexes_color,
            linewidths=0.5
        );

        pd.DataFrame(point_cloud).rename(columns={0: 'x', 1: 'y'}).plot(
            kind='scatter',
            x='x',
            y='y',
            ax=axs[idx],
            s=7,
            c=one_simplexes_color
        );
        
        axs[idx].set_title('epsilon = {EPSILON}'.format(EPSILON=epsilon))

        axs[idx].add_collection(lc2);
        axs[idx].autoscale();

        for two_simplex in two_simplicies:
            Path = mpath.Path
            path_data = [
                    (Path.MOVETO, point_cloud[two_simplex[0]]),
                    (Path.LINETO, point_cloud[two_simplex[1]]),
                    (Path.LINETO, point_cloud[two_simplex[2]]),
                    (Path.CLOSEPOLY, point_cloud[two_simplex[0]]),
                    ]
            codes, verts = zip(*path_data)
            path = mpath.Path(verts, codes)
            patch = mpatches.PathPatch(path, facecolor=two_simplexes_color, alpha=0.2);
            axs[idx].add_patch(patch);

plt.show();
```

# Toy examples

## Noisy circle

```python
n_pts = 50
rrnd = 0.5
alpha = [0.4, 0.6, 0.2, 0.4]

pc = [[np.cos(x*np.pi) + alpha[0]*random.random(), np.sin(x*np.pi) + alpha[1]*random.random()] for 
      x in [(r + random.random()*rrnd)*(1/float(n_pts)) for r in range(-n_pts, n_pts)]]

dm = distance_matrix(pc, pc)
```

```python
plot_two_simplexes_multi(point_cloud=pc, distance_matrix=dm, epsilon_list=[0.2, 0.3, 0.45], one_simplexes_color='#6DA5BA', two_simplexes_color='#E0513B')
```

```python
h = point_cloud_to_persistent_homology(point_cloud=pc, epsilon_range=np.array([x*0.01 for x in range(100)]), max_simplex_dim=2)
```

```python
pd.DataFrame(h).rename(columns={0: 'h0', 1: 'h1'}).plot(
    kind='line',
    lw=1,
    figsize=(20, 5),
    secondary_y='h1'
);
```

# Bandpass filtering & circular parametrization

```python
from scipy.signal import butter, lfilter

def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a


def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y
```

```python
fs = 2000.0
T = 0.05
nsamples = int(T * fs)
t = np.linspace(0, T, nsamples, endpoint=False)
a = 0.02
f0 = 600.0
x = 0.1 * np.sin(2 * np.pi * 1.2 * np.sqrt(t))
x += 0.01 * np.cos(2 * np.pi * 312 * t + 0.1)
x += a * np.cos(2 * np.pi * f0 * t + .11)
x += 0.03 * np.cos(2 * np.pi * 2000 * t)

y = butter_bandpass_filter(data=x, lowcut=500.0, highcut=1250.0, fs=5000.0, order=10)

color_original = '#3BBBE0'
color_bandpass = '#E0983B'

pd.DataFrame([x, y]).T.rename(columns={
    0: 'original',
    1: 'bandpass_filtered'
}).plot(
    kind='line',
    lw=1,
    figsize=(17, 5),
    color=[color_original, color_bandpass]
);

fig1, axs1 = plt.subplots(1, 2, figsize=(14, 7));

pd.DataFrame(time_series_to_point_cloud(time_series=x, dimension_embed=2)).plot(
    kind='scatter',
    x=0,
    y=1,
    ax=axs1[0],
    s=5,
    alpha=0.5,
    title='original',
    c=color_original
);

pd.DataFrame(time_series_to_point_cloud(time_series=y, dimension_embed=2)).plot(
    kind='scatter',
    x=0,
    y=1,
    ax=axs1[1],
    s=5,
    alpha=0.5,
    title='bandpass_filtered',
    c=color_bandpass
);

plt.tight_layout();


fig2, axs2 = plt.subplots(1, 2, figsize=(14, 7));
axs2[0].set_title(label='persistent homology original')
axs2[1].set_title(label='persistent homology bandpass filtered')

# build filtration with fill_rips
filtration_original = d.fill_rips(time_series_to_point_cloud(x), 2, 1.0)

# calculate persistent homology
persistence_original = d.homology_persistence(filtration_original)

# initialize persistence diagrams
diagrams_original = d.init_diagrams(persistence_original, filtration_original)

# plt.figure(figsize=(7, 7))
# d.plot.plot_bars(diagrams[1], show=True);

maxdeath = max([point.death for point in diagrams_original[1]])

for index, point in enumerate(diagrams_original[1]):
    axs2[0].axhline(y=index, xmin=point.birth/maxdeath, xmax=point.death/maxdeath, color=color_original, linestyle='-', lw=1);

# plt.show();






# build filtration with fill_rips
filtration_bandpass = d.fill_rips(time_series_to_point_cloud(y), 2, 1.0)

# calculate persistent homology
persistence_bandpass = d.homology_persistence(filtration_bandpass)

# initialize persistence diagrams
diagrams_bandpass = d.init_diagrams(persistence_bandpass, filtration_bandpass)

# plt.figure(figsize=(7, 7))
# d.plot.plot_bars(diagrams[1], show=True);

maxdeath = max([point.death for point in diagrams_bandpass[1]])

for index, point in enumerate(diagrams_bandpass[1]):
    axs2[1].axhline(y=index, xmin=point.birth/maxdeath, xmax=point.death/maxdeath, color=color_bandpass, linestyle='-', lw=1);

plt.tight_layout();
plt.show();


# circular parametrization
fig3, axs3 = plt.subplots(1, 2, figsize=(14, 7));
axs3[0].set_title(label='circular parametrization scatter original')
axs3[1].set_title(label='circular parametrization scatter bandpass')


prime = 11

co_persistence_original = d.cohomology_persistence(filtration_original, prime, True)
co_diagrams_original = d.init_diagrams(co_persistence_original, filtration_original)

longest_cohomology_class_original = max(co_diagrams_original[1], key = lambda pt: pt.death - pt.birth)
cocycle_original = p.cocycle(longest_cohomology_class_original.data)

filtration_restricted_original = d.Filtration([s for s in filtration_original if s.data <= (longest_cohomology_class_original.death + longest_cohomology_class_original.birth)/2])
vertex_values_original = d.smooth(filtration_restricted_original, cocycle_original, prime)


# axs3[0].scatter(time_series_to_point_cloud(x)[:,0], time_series_to_point_cloud(x)[:,1], c=vertex_values_original, cmap='hsv', s=5, alpha=0.5);


co_persistence_bandpass = d.cohomology_persistence(filtration_bandpass, prime, True)
co_diagrams_bandpass = d.init_diagrams(co_persistence_bandpass, filtration_bandpass)

longest_cohomology_class_bandpass = max(co_diagrams_bandpass[1], key = lambda pt: pt.death - pt.birth)
cocycle_bandpass = p.cocycle(longest_cohomology_class_bandpass.data)

filtration_restricted_bandpass = d.Filtration([s for s in filtration_bandpass if s.data <= (longest_cohomology_class_bandpass.death + longest_cohomology_class_bandpass.birth)/2])
vertex_values_bandpass = d.smooth(filtration_restricted_bandpass, cocycle_bandpass, prime)


# axs3[1].scatter(time_series_to_point_cloud(y)[:,0], time_series_to_point_cloud(y)[:,1], c=vertex_values_bandpass, cmap='hsv', s=5, alpha=0.5);

pd.DataFrame(time_series_to_point_cloud(time_series=x, dimension_embed=2)).plot(
    kind='scatter',
    x=0,
    y=1,
    ax=axs3[0],
    s=5,
    alpha=0.8,
    title='original',
    c=vertex_values_original,
    cmap='plasma',
    colorbar=False
);

pd.DataFrame(time_series_to_point_cloud(time_series=y, dimension_embed=2)).plot(
    kind='scatter',
    x=0,
    y=1,
    ax=axs3[1],
    s=5,
    alpha=0.8,
    title='bandpass_filtered',
    c=vertex_values_bandpass,
    cmap='plasma',
    colorbar=False
);

plt.tight_layout();


original_data = pd.DataFrame(x)
original_data.reset_index(inplace=True)
original_data.rename(columns={'index': 'x', 0: 'y'}, inplace=True)
original_data['vertex_values'] = pd.Series(data=vertex_values_original, index=original_data.index[1:])

bandpass_data = pd.DataFrame(y)
bandpass_data.reset_index(inplace=True)
bandpass_data.rename(columns={'index': 'x', 0: 'y'}, inplace=True)
bandpass_data['vertex_values'] = pd.Series(data=vertex_values_bandpass, index=bandpass_data.index[1:])

fig4, ax4 = plt.subplots(figsize=(20, 5));

original_data[[
    'x',
    'y'
]].rename(columns={'y': 'original'}).plot(
    kind='line',
    x='x',
    ax=ax4,
    lw=0.5,
    color=color_original
);


original_data.plot(
    kind='scatter',
    x='x',
    y='y',
    ax=ax4,
    c=original_data['vertex_values'],
    cmap='plasma',
    s=7,
    alpha=0.7,
    colorbar=False
);

bandpass_data[[
    'x',
    'y'
]].rename(columns={'y': 'bandpass_filtered'}).plot(
    kind='line',
    x='x',
    ax=ax4,
    lw=0.5,
    color=color_bandpass
);

bandpass_data.plot(
    kind='scatter',
    x='x',
    y='y',
    ax=ax4,
    c=bandpass_data['vertex_values'],
    cmap='plasma',
    s=7,
    alpha=0.7,
    colorbar=False
);



fig5, ax5 = plt.subplots(figsize=(20, 5));


bt_sum_original = time_series_rolling_betti(time_series=original_data['y']/max(original_data['y']),
                                   epsilon_range=np.array([y*.01 for y in range(100)]),
                                   max_simplex_dim=3,
                                   window=20,
                                   agg='sum')

bt_sum_original.rename(columns={'betti': 'betti_sum_original', 'y': 'original'}, inplace=True)

bt_sum_original['original'] = bt_sum_original['original']*max(original_data['y'])

bt_sum_original.plot(
    kind='line',
    lw=1,
    figsize=(20, 5),
    secondary_y='betti_sum_original',
    ax=ax5
);

bt_sum_bandpass = time_series_rolling_betti(time_series=bandpass_data['y']/max(bandpass_data['y']),
                                   epsilon_range=np.array([y*.01 for y in range(100)]),
                                   max_simplex_dim=3,
                                   window=20,
                                   agg='sum')

bt_sum_bandpass.rename(columns={'betti': 'betti_sum_bandpass', 'y': 'bandpass_filtered'}, inplace=True)

bt_sum_bandpass['bandpass_filtered'] = bt_sum_bandpass['bandpass_filtered']*max(bandpass_data['y'])

bt_sum_bandpass.plot(
    kind='line',
    lw=1,
    figsize=(20, 5),
    secondary_y='betti_sum_bandpass',
    ax=ax5
);
```
